# Analyze MCMC chain with GetDist

```python
# show plots inline
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
# load main getdist plot module and samples class
from getdist import plots, MCSamples
import getdist
# use this *after* importing getdist if you want to use interactive plots
# %matplotlib notebook
import matplotlib.pyplot as plt
import IPython
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
# matplotlib 2 may not work very well without usetex on, can uncomment
# plt.rcParams['text.usetex']=True
```

```python
# load chain
samples_KV_H0_Dz_IA = getdist.loadMCSamples('./KV450_H0_Dz_IA/KV450_H0_Dz_IA')
samples_KV_H1_Dz_IA = getdist.loadMCSamples('./KV450_H1_Dz_IA/KV450_H1_Dz_IA')
samples_P_H0_Dz_IA = getdist.loadMCSamples('./Planck_H0_Dz_IA/Planck_H0_Dz_IA')
samples_P_H1_Dz_IA = getdist.loadMCSamples('./Planck_H1_Dz_IA/Planck_H1_Dz_IA')
```

```python
# Triangle plot
# H0
g = plots.get_subplot_plotter()
g.settings.axis_tick_x_rotation=45
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.8
g.settings.title_limit_fontsize = 16
g.settings.legend_fontsize = 16
g.settings.axes_fontsize=16
g.settings.axes_labelsize=16
g.triangle_plot([samples_KV_H0_Dz_IA, samples_P_H0_Dz_IA], 
                param_limits = {'A_IA':[-6,6], 
                                'D_z1':[-0.117, 0.117], 
                                'D_z2':[-0.069, 0.069],
                                'D_z3':[-0.078, 0.078], 
                                'D_z4':[-0.036, 0.036],
                                'D_z5':[-0.033, 0.033]
                               },
    filled=True, 
    legend_labels=['KV450 cosmology', 'Planck cosmology'], 
    legend_loc='upper right', 
    contour_colors=['red','black'],
#     title_limit=1 # first title limit (for 1D plots) is 68% by default
    )

plt.savefig('./MCMC_H0.pdf', dpi=300)

# H1
g = plots.get_subplot_plotter()
g.settings.axis_tick_x_rotation=45
g.settings.figure_legend_frame = False
g.settings.alpha_filled_add=0.8
g.settings.title_limit_fontsize = 16
g.settings.legend_fontsize = 16
g.settings.axes_fontsize=16
g.settings.axes_labelsize=16
g.triangle_plot([samples_KV_H1_Dz_IA, samples_P_H1_Dz_IA], 
                param_limits = {'A_IA_m':[-6,6],
                                'A_IA_s':[-6,6],
                                'D_z1_m':[-0.117, 0.117], 
                                'D_z2_m':[-0.069, 0.069],
                                'D_z3_m':[-0.078, 0.078], 
                                'D_z4_m':[-0.036, 0.036],
                                'D_z5_m':[-0.033, 0.033],
                                'D_z1_s':[-0.3, 0.3], 
                                'D_z2_s':[-0.3, 0.3],
                                'D_z3_s':[-0.3, 0.3], 
                                'D_z4_s':[-0.3, 0.3],
                                'D_z5_s':[-0.3, 0.3]
                               },                
                filled=True, 
                legend_labels=['KV450 cosmology', 'Planck cosmology'], 
                legend_loc='upper right', 
                contour_colors=['red', 'black'],
#                 title_limit=1, # first title limit (for 1D plots) is 68% by default
                markers = {'A_IA_s':0,
                            'D_z1_s':0, 
                            'D_z2_s':0,
                            'D_z3_s':0, 
                            'D_z4_s':0,
                            'D_z5_s':0},
                marker_args = {'color':'gray', 
                               'ls':'--', 
                               'lw': 2}
    )
plt.savefig('./MCMC_H1.pdf', dpi=300)
```

```python
# Mean value table
print(samples_KV_H0_Dz_IA.getTable(limit=1).tableTex())
print(samples_KV_H1_Dz_IA.getTable(limit=1).tableTex())
# Mean value table
print(samples_P_H0_Dz_IA.getTable(limit=1).tableTex())
print(samples_P_H1_Dz_IA.getTable(limit=1).tableTex())
```

```python
# simple numpy analysis
#
import numpy as np
data_KV_H0 = np.loadtxt("./KV450_H0_Dz_IA/KV450_H0_Dz_IA.txt")
data_KV_H1 = np.loadtxt("./KV450_H1_Dz_IA/KV450_H1_Dz_IA.txt")
data_P_H0 = np.loadtxt("./Planck_H0_Dz_IA/Planck_H0_Dz_IA.txt")
data_P_H1 = np.loadtxt("./Planck_H1_Dz_IA/Planck_H1_Dz_IA.txt")

# min chi2
for data in [data_KV_H0, data_KV_H1, data_P_H0, data_P_H1]:
    max_index = np.argmin(data[:,1])
    print(data[max_index])
    print("chi^2", data[max_index,1]*2.)
```
