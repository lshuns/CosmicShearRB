# Analyze downloaded MCMC chain with GetDist

```python
# show plots inline
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
# load main getdist plot module and samples class
from getdist import plots, MCSamples
import getdist
# use this *after* importing getdist if you want to use interactive plots
# %matplotlib notebook
import matplotlib.pyplot as plt
import IPython
print('GetDist Version: %s, Matplotlib version: %s'%(getdist.__version__, plt.matplotlib.__version__))
# matplotlib 2 may not work very well without usetex on, can uncomment
# plt.rcParams['text.usetex']=True
```

```python
# load chain
samples = getdist.loadMCSamples('./KV450_fiducial/KV450_fiducial')
```

```python
# triangle plot
# omegam vs. sigma8
g1 = getdist.plots.get_single_plotter()
g1.plot_2d(samples, ['omegam', 'sigma8'], filled=True)
# omegam vs. S8
g2 = getdist.plots.get_single_plotter()
g2.plot_2d(samples, ['omegam', 'S8'], filled=True)
```

```python
print(samples.getTable().tableTex())
```

```python
# simple numpy analysis
import numpy as np
data = np.loadtxt("./KV450_fiducial/KV450_fiducial.txt")
```

```python
# maximum likelihood parameters
max_index = np.argmax(data[:,1]*data[:,0])
print(data[max_index])
print("chi^2", data[max_index,1]*2.)
```
